```python
def plot_df_comparison(*dfs, col_num, real_value, save_path=None):
    """Plot estimation error with error bars (sorted by numeric sample size)."""

    # ---------- Gather data ----------
    sample_sizes = [250 * (2 ** i) for i in range(len(dfs))]
    data_list = []

    for N, df in zip(sample_sizes, dfs):
        # FIX: select by column position, not column name
        col_values = df.iloc[:, col_num].to_numpy() - real_value
        data_list.append(pd.DataFrame({"value": col_values, "N": N}))

    data = pd.concat(data_list, ignore_index=True)

    # ---------- Compute summary ----------
    summary = (
        data.groupby("N")["value"]
        .agg(["mean", "std", "count"])
        .sort_index()
    )
    summary["se"] = summary["std"] / np.sqrt(summary["count"])  # FIX: SE

    x_positions = np.arange(len(summary))

    # ---------- Plot ----------
    plt.figure(figsize=(10, 5))

    sns.stripplot(
        x="N", y="value", data=data,
        color="black", size=3, jitter=True, alpha=0.6,
        order=summary.index
    )

    for i, (N, row) in enumerate(summary.iterrows()):
        plt.errorbar(
            x=i, y=row["mean"], yerr=row["se"],
            fmt="none", ecolor="red", elinewidth=3,
            capsize=8, capthick=2.5, alpha=0.9, zorder=5
        )
        plt.plot(
            i, row["mean"], marker="o", color="blue",
            markersize=6, markeredgecolor="black",
            markeredgewidth=0.5, zorder=6
        )

    if col_num == 0:
        theta = r"$\sigma^2_{g \times g}$"
    elif col_num == 1:
        theta = r"$\sigma^2_{e}$"
    else:
        theta = "Parameter"

    plt.axhline(0, color="gray", linestyle="--", linewidth=1)
    plt.title(
        f"Error of {theta} by Sample Size (real value = {real_value})",
        fontsize=14, pad=10
    )
    plt.xlabel("Sample Size (N)", fontsize=12)
    plt.ylabel(f"Error: {theta} - {real_value}", fontsize=12)
    plt.xticks(ticks=x_positions, labels=[f"N={N}" for N in summary.index], fontsize=11)
    plt.tight_layout()

    if save_path:
        os.makedirs(os.path.dirname(save_path), exist_ok=True)
        plt.savefig(save_path, dpi=600, bbox_inches="tight")
        print(f"Plot saved to: {save_path}")
    else:
        plt.show()
```

```python
### a toy exmaple to show the unbiasness of MoM and the Se change with individual(random SNP case)
def run_mc(X, n_mc=40, s2gxg=0.9, s2e=0.1):
    gxgs = []
    es = []
    
    for _ in range(n_mc):
        Z, y = simulate_from_raw_only_W(X, s2gxg=s2gxg, s2e=s2e)
        gxg, e, _ = MoM_only_M(Z, y, nmc=40)
        gxgs.append(gxg)
        es.append(e)
    
    gxgs = np.array(gxgs).flatten()
    es = np.array(es).flatten()
    
    return pd.DataFrame({"gxgs": gxgs, "es": es})


# Run for individual size
n_mc = 100
X1 = RandomSNP[:250, :100]
X2 = RandomSNP[:500, :100]
X3 = RandomSNP[:1000, :100]

df1 = run_mc(X1, n_mc)
df2 = run_mc(X2, n_mc)
df3 = run_mc(X3, n_mc)
```

```python
plot_df_comparison(df1, df2, df3,  col_num=0, real_value=0.9, save_path=False)
```
