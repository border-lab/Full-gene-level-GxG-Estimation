```python
### a toy exmaple to show the unbiasness of MoM and the Se change with individual(random SNP case)
def run_experiment_MC(X, n_mc=100, s2gxg=0.9, s2e=0.1):
    """
    Run Monte Carlo simulations to estimate GxG and residual variance components.
    X: (n, m) raw genotype data
    n_mc: number of Monte Carlo replicates
    """
    gxgs = []
    es = []
    
    for _ in range(n_mc):
        Z, y = simulate_from_raw_only_W(X, s2gxg=0.9, s2e=0.1)
        gxg, e, _ = MoM_only_M(Z, y, nmc=40)
        gxgs.append(gxg)
        es.append(e)
    
    gxgs = np.array(gxgs).flatten()
    es = np.array(es).flatten()
    
    return pd.DataFrame({"gxgs": gxgs, "es": es})

# Run for individual size
n_mc = 40
X1 = RandomSNP[:250, :100]
X2 = RandomSNP[:500, :100]
X3 = RandomSNP[:1000, :100]

df1 = run_experiment_MC(X1, n_mc)
df2 = run_experiment_MC(X2, n_mc)
df3 = run_experiment_MC(X3, n_mc)
```

```python

plot_relative_error_accross_sample_size(
    df1, df2, df3, 
    basic_individual=250, 
    col_num=0, 
    real_value=0.9, 
    save_path=None
)
```
