```python
def plot_distribution(data, bins=100, target=1.0, title_prefix="Distribution"):
    mu = np.mean(data)
    std = np.std(data)
    plt.figure(figsize=(10, 6))
    plt.hist(data, bins=bins, alpha=0.7, color='skyblue', edgecolor='white')
    plt.axvline(x=mu, color='black', linewidth=2, label=f'Mean = {mu:.3f}')
    plt.axvline(x=target, color='red', linestyle='--', linewidth=2, label=f'Target = {target}')
    plt.title(f'{title_prefix}: μ={mu:.3f}, σ={std:.3f}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(axis='y', alpha=0.3)
    
    plt.show()
```

```python
#### why the last week's method is failed
n, m = Z.shape

p = m * (m - 1) // 2
A = np.zeros((p, m))
b = np.zeros(p)

row = 0
for i in range(m):
    for j in range(i + 1, m):
        A[row, i] = 1
        A[row, j] = 1
        v = (Z[:, i] * Z[:, j]).var()
        b[row] = -0.5 * np.log(max(v, 1e-10))
        row += 1

# Compute LS solution
u_ls, *_ = np.linalg.lstsq(A, b, rcond=None)
w_ls = np.exp(u_ls)  # Convert to weights

# Objective 1: Log-space 
def log_space_objective_w(w):
    """Log-space objective using weights w"""
    u = np.log(w)  # Convert w to u
    total = 0
    for i in range(m):
        for j in range(i + 1, m):
            v = (Z[:, i] * Z[:, j]).var()
            c_ij = -0.5 * np.log(max(v, 1e-10))
            total += (u[i] + u[j] - c_ij) ** 2
    return total

# Objective 2: Original-space 
def original_space_objective_w(w):
    """Original-space objective using weights w"""
    Zw = Z * w 
    total = 0
    for i in range(m):
        for j in range(i + 1, m):
            v = (Zw[:, i] * Zw[:, j]).var()
            total += (v - 1) ** 2
    return total

# Define weights
w_unweighted = np.ones(m)  # w = 1

print("=== Log-space objective ===")
print(f"Unweighted (w=1): {log_space_objective_w(w_unweighted):.4f}")
print(f"LS weights:       {log_space_objective_w(w_ls):.4f}")

print("\n=== Original-space objective ===")
print(f"Unweighted (w=1): {original_space_objective_w(w_unweighted):.4f}")
print(f"LS weights:       {original_space_objective_w(w_ls):.4f}")

print("\n" + "="*60)
print("CONCLUSION:")
print("- LS minimizes Log-space objective, but NOT Original-space")
print("="*60)
```

```python
#### use gradient descent method

def original_space_optimization(Z, max_iter=1000, tol=1e-8):
    """
    Minimize sum_{i<j} (Var(w_i w_j Z_i Z_j) - 1)^2 directly.
    
    Parameters
    ----------
    Z : ndarray of shape (n, m)
        Standardized genotype matrix
    max_iter : int
        Maximum iterations
    tol : float
        Gradient tolerance for convergence
    
    Returns
    -------
    w_opt : ndarray of shape (m,)
        Optimal weights
    u_opt : ndarray of shape (m,)
        Optimal log-weights
    result : OptimizeResult
        Optimization result object
    """
    n, m = Z.shape
    # Estimate allele frequencies p_i from E[X_i] = 2*p_i (Hardy-Weinberg)
    p_i = np.clip(X.mean(axis=0) / 2, 0.01, 0.99)

    # Estimate covariance matrix of genotypes
    sigma_ij = np.cov(X, rowvar=False)

    def var_ZiZj(pi, pj, sij):
        """Compute Var(Z_i * Z_j) given allele frequencies and covariance."""
        numerator = sij * (1 - 2*pi) * (1 - 2*pj)
        denominator = 4 * pi * pj * (1 - pi) * (1 - pj)
        return 1 + numerator / denominator

    
    # Precompute V_ij = Var(Z_i Z_j) for all pairs
    V = np.zeros((m, m))
    for i in range(m):
        for j in range(i + 1, m):
            V[i, j] = (Z[:, i] * Z[:, j]).var()
            V[j, i] = V[i, j]
    
    def objective(u):
        """f(u) = sum_{i<j} (exp(2u_i + 2u_j) * V_ij - 1)^2"""
        total = 0.0
        for i in range(m):
            for j in range(i + 1, m):
                r_ij = np.exp(2*u[i] + 2*u[j]) * V[i, j] - 1
                total += r_ij ** 2
        return total
    
    def gradient(u):
        """Gradient of f with respect to u"""
        grad = np.zeros(m)
        for k in range(m):
            for j in range(m):
                if j != k:
                    i_idx, j_idx = min(k, j), max(k, j)
                    exp_term = np.exp(2*u[k] + 2*u[j])
                    V_kj = V[i_idx, j_idx]
                    r_kj = exp_term * V_kj - 1
                    grad[k] += 4 * r_kj * exp_term * V_kj
        return grad
    
    # Initialize with u = 0 (w = 1)
    u0 = np.zeros(m)
    
    # Optimize using L-BFGS-B
    result = minimize(
        objective,
        u0,
        method='L-BFGS-B',
        jac=gradient,
        options={'maxiter': max_iter, 'gtol': tol}
    )
    
    u_opt = result.x
    w_opt = np.exp(u_opt)
    
    return w_opt, u_opt, result
```

```python
#### filter out the SNP under MAF thres
def filter_maf(G, maf_min=0.05):
    
    af = np.nanmean(G, axis=0) / 2.0
    maf = np.minimum(af, 1 - af)
    
    # Keep variants with MAF >= threshold
    keep_mask = maf >= maf_min

    # Return filtered data in its original format
    if isinstance(G, pd.DataFrame):
        G_filtered = G.loc[:, keep_mask]
    else:
        G_filtered = G[:, keep_mask]
    
    return G_filtered, maf, keep_mask
```

```python
###check the high LD pair
#X1 = RandomSNP[:4000,:600] 
X1 = ContiguousSNP[:2000,:400]
X1 = filter_maf(X1)[0]
Z = (X1 - X1.mean(axis=0)) / X1.std(axis=0)
n, m = Z.shape
w_unweighted = np.ones(m)  
def check_weighted_products_with_indices(Z, w=None, min_dist=1, var_threshold=2.0):
    """
    Check variance of weighted products and return indices of high variance pairs.
    
    Parameters
    ----------
    Z : ndarray of shape (n, m)
        Standardized genotype matrix
    w : ndarray of shape (m,), optional
        Weights (if None, uses w=1)
    min_dist : int
        Minimum distance |j - i| to include
    var_threshold : float
        Threshold to identify high variance pairs
    
    Returns
    -------
    var_matrix : ndarray of shape (m, m)
        Variance matrix
    stats : dict
        Statistics
    high_var_pairs : list of tuples
        List of (i, j, variance) for pairs with variance > threshold
    """
    n, m = Z.shape
    
    if w is None:
        w = np.ones(m)
    
    Zw = Z * w
    
    # Compute variance matrix
    var_matrix = np.zeros((m, m))
    high_var_pairs = []
    all_pairs = []
    
    for i in range(m):
        for j in range(i + min_dist, m):
            v = (Zw[:, i] * Zw[:, j]).var()
            var_matrix[i, j] = v
            var_matrix[j, i] = v
            all_pairs.append((i, j, v))
            
            if v > var_threshold:
                high_var_pairs.append((i, j, v))
    
    # Sort high variance pairs by variance (descending)
    high_var_pairs.sort(key=lambda x: x[2], reverse=True)
    
    # Compute statistics
    all_vars = np.array([p[2] for p in all_pairs])
    stats = {
        'mean': all_vars.mean(),
        'std': all_vars.std(),
        'min': all_vars.min(),
        'max': all_vars.max(),
        'n_pairs': len(all_pairs),
        'n_high_var': len(high_var_pairs),
    }
    
    return var_matrix, stats, high_var_pairs, all_pairs


# Usage
var_matrix, stats, high_var_pairs, all_pairs = check_weighted_products_with_indices(
    Z, w=np.ones(m), min_dist=1, var_threshold=2.0
)
print(f"=== Statistics ===")
print(f"Mean: {stats['mean']:.4f}")
print(f"Std:  {stats['std']:.4f}")
print(f"Max:  {stats['max']:.4f}")
print(f"Min:  {stats['min']:.4f}")
print(f"Total pairs: {stats['n_pairs']}")
print(f"High variance pairs (> 2.0): {stats['n_high_var']}")

print(f"\n=== Top 20 High Variance Pairs ===")
print(f"{'Rank':<6} {'SNP i':<8} {'SNP j':<8} {'Distance':<10} {'Var(Zi Zj)':<12}")
print("-" * 50)
for rank, (i, j, v) in enumerate(high_var_pairs[:20], 1):
    print(f"{rank:<6} {i:<8} {j:<8} {j-i:<10} {v:<12.4f}")
```

```python
def original_space_optimization_symmetric(Z, max_iter=1000, tol=1e-8, penalty=10000.0):
    """
    Minimize variance of Var(w_i w_j Z_i Z_j) distribution
    subject to: mean(Var(w_i w_j Z_i Z_j)) = 1
    
    This produces a symmetric, tight distribution around 1.
    """
    n, m = Z.shape
    n_pairs = m * (m - 1) // 2
    
    # Precompute V_ij = Var(Z_i Z_j) for all pairs
    V = np.zeros((m, m))
    for i in range(m):
        for j in range(i + 1, m):
            V[i, j] = (Z[:, i] * Z[:, j]).var()
            V[j, i] = V[i, j]
    
    def objective(u):
        """
        Minimize: Var(V_ij^w) + penalty * (Mean(V_ij^w) - 1)^2
        
        Var(V) = E[V^2] - E[V]^2 = mean(V^2) - mean(V)^2
        """
        var_sum = 0.0
        var_sq_sum = 0.0
        
        for i in range(m):
            for j in range(i + 1, m):
                var_ij = np.exp(2*u[i] + 2*u[j]) * V[i, j]
                var_sum += var_ij
                var_sq_sum += var_ij ** 2
        
        mean_var = var_sum / n_pairs
        mean_var_sq = var_sq_sum / n_pairs
        variance_of_var = mean_var_sq - mean_var ** 2  # Var = E[X^2] - E[X]^2
        
        # Objective: minimize variance + penalty for mean != 1
        return variance_of_var + penalty * (mean_var - 1) ** 2
    
    def gradient(u):
        """Gradient of the objective"""
        grad = np.zeros(m)
        
        var_sum = 0.0
        var_sq_sum = 0.0
        
        # First pass: compute mean and mean_sq
        for i in range(m):
            for j in range(i + 1, m):
                var_ij = np.exp(2*u[i] + 2*u[j]) * V[i, j]
                var_sum += var_ij
                var_sq_sum += var_ij ** 2
        
        mean_var = var_sum / n_pairs
        mean_var_sq = var_sq_sum / n_pairs
        
        # Second pass: compute gradients
        for k in range(m):
            d_mean_du_k = 0.0
            d_mean_sq_du_k = 0.0
            
            for j in range(m):
                if j != k:
                    i_idx, j_idx = min(k, j), max(k, j)
                    exp_term = np.exp(2*u[k] + 2*u[j])
                    V_kj = V[i_idx, j_idx]
                    var_kj = exp_term * V_kj
                    
                    # d(var_kj)/du_k = 2 * var_kj
                    d_var_kj = 2 * var_kj
                    
                    # d(var_kj^2)/du_k = 2 * var_kj * d(var_kj)/du_k = 4 * var_kj^2
                    d_var_kj_sq = 4 * var_kj ** 2
                    
                    d_mean_du_k += d_var_kj / n_pairs
                    d_mean_sq_du_k += d_var_kj_sq / n_pairs
            
            # d(Variance)/du_k = d(mean_sq)/du_k - 2 * mean * d(mean)/du_k
            d_variance_du_k = d_mean_sq_du_k - 2 * mean_var * d_mean_du_k
            
            # d(penalty)/du_k = 2 * penalty * (mean - 1) * d(mean)/du_k
            d_penalty_du_k = 2 * penalty * (mean_var - 1) * d_mean_du_k
            
            grad[k] = d_variance_du_k + d_penalty_du_k
        
        return grad
    
    # Initialize with u = 0 (w = 1)
    u0 = np.zeros(m)
    
    # Optimize using L-BFGS-B
    result = minimize(
        objective,
        u0,
        method='L-BFGS-B',
        jac=gradient,
        options={'maxiter': max_iter, 'gtol': tol}
    )
    
    u_opt = result.x
    w_opt = np.exp(u_opt)
    
    return w_opt, u_opt, result
    
def weighted_estimation_pipeline(X, maf_threshold=0.05, penalty=10000.0, max_iter=1000):
    """
    Complete pipeline for variance-corrected epistatic estimation.
    
    Steps:
    1. Filter low MAF SNPs
    2. Standardize genotypes
    3. Compute optimal weights (minimize variance with mean=1 constraint)
    4. Apply weights
    
    Parameters
    ----------
    X : ndarray of shape (n, m)
        Raw genotype matrix (values in {0, 1, 2})
    maf_threshold : float
        Minimum MAF threshold
    penalty : float
        Penalty for mean constraint
    max_iter : int
        Maximum optimization iterations
    
    Returns
    -------
    Z_weighted : ndarray
        Weighted standardized genotype matrix
    w : ndarray
        Optimal weights
    """
    
    n, m_orig = X.shape
    
    # Compute MAF
    p = X.mean(axis=0) / 2
    maf = np.minimum(p, 1 - p)
    keep_idx = maf >= maf_threshold
    
    X_filtered = X[:, keep_idx]
    m = X_filtered.shape[1]
    Z = (X_filtered - X_filtered.mean(axis=0)) / X_filtered.std(axis=0)   
    w, u, result = original_space_optimization_symmetric(Z, max_iter=max_iter, penalty=penalty)
    Z_weighted = Z * w
    

    return Z_weighted, w
```
